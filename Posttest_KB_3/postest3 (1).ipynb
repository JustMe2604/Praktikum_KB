```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Membaca dan Menampilkan File CSV

```python
data = pd.read_csv("Airplane_Crashes_and_Fatalities_Since_1908.csv")
data.head()
```

# Menampilkan Informasi Dataset

```python
data.info()
```

```python
data.describe()
```

# Menangani Missing Value pada kolom Index,Aboard,Fatalities dan Ground

```python
index = (data.index == 0).sum()
Aboard = (data.Aboard == 0).sum()
Fatalities = (data.Fatalities == 0).sum()
Ground = (data.Ground == 0).sum()

 
print("Nilai 0 di kolom Index ada: ", index)
print("Nilai 0 di kolom Aboard ada: ", Aboard)
print("Nilai 0 di kolom Fatalities ada: ", Fatalities)
print("Nilai 0 di kolom Ground ada: ", Ground)
```

```python
data.loc[(data['index']==0)]
data.loc[(data['Aboard']==0)]
data.loc[(data['Fatalities']==0)]
data.loc[(data['Ground']==0)]
```

```python
data.describe()
```

# Memvisualisasikan Data menggunakan Diagram Garis

```python
x = data.Aboard[0:5]
y = data.index[0:5]
plt.figure(figsize = (5, 3))
plt.title("Contoh Grafik")
plt.xlabel("aboard")
plt.ylabel("Index")
plt.plot(x, label = "aboard", color = "green")
plt.plot(y, label = "Index", marker = "o")
plt.legend(loc = "upper right")
plt.show()
```

# Memvisualisasikan Data menggunakan Diagram Batang (Histogram)

```python
x = list(data.Time[0:15].astype(str))
y = list(data.Time[0:15].astype(str))
plt.bar(x,y, color="purple")
plt.xlabel("Nilai X")
plt.ylabel("Nilai Y")
plt.title("Perbandingan Waktu")
plt.show()
```

# Memvisualisasikan Data Menggunakan Grafik Heatmap

```python
fig, ax = plt.subplots(figsize=(10,6))
sns.heatmap(data.corr(), center=0, cmap='Blues')
ax.set_title('Airplane_Crashes_and_Fatalities_Since_1908.csv')
ax.set_xlabel("Nilai X")
ax.set_ylabel("Nilai Y")
```

# Grafik Pie

```python
Grafik = data['Aboard'].value_counts().rename_axis('Aboard').reset_index(name='Id_count')
plt.figure(figsize=(20, 20))
plt.pie(Grafik.Id_count,labels = Grafik.Aboard, startangle=90, autopct="%.1f%%")
plt.legend()
```
